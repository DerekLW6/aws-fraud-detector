## Important - You have to convert the time columns to EVENT_Timestamp it's required for AWS Fraud Detector. I did it in excel prior to writing this notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
%matplotlib inline

# importing the data
df = pd.read_csv('creditcard.csv')

# Renaming
df.rename(columns = {'Class':'EVENT_LABEL'}, inplace = True)

#Lowercase
df.columns = map(str.lower, df.columns)

# Fixing two outliers
df.rename(columns = {'event_label':'EVENT_LABEL', 'event_timestamp':'EVENT_TIMESTAMP'}, inplace = True)

df.head(3)
```

```python
# Determining the distribution
sns.countplot(x=df['EVENT_LABEL'], data=df)
```

```python
df['EVENT_LABEL'].value_counts(normalize=True)
```

```python
# Replacing labels
df['EVENT_LABEL'] = df['EVENT_LABEL'].replace(0,'legit')
df['EVENT_LABEL'] = df['EVENT_LABEL'].replace(1,'fraud')
```

```python
# Writting to CSV
df.to_csv('fraud.csv')
```
